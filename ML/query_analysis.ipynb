```python
# imports, load data

import pandas as pd
import matplotlib.pyplot as plt
import nltk
import numpy as np

nltk.download('punkt')

with open('/home/cadu/Downloads/livros/Modern Engineering Statistics/resources/queries/all_queries') as f:
    lines = f.read().splitlines() 
```

```python
# create a dictionary with size, [query lists]
query_size_results = {}

for term in lines:
    query_term = str(term)
    size = len(nltk.word_tokenize(query_term))
    
    if size not in query_size_results.keys():
        query_size_results[size] = [query_term.replace("de", "").replace("para", "").replace(".", "").replace("com", "").replace(" e ", "")]
    else:
        query_list = query_size_results[size]
        query_list.append(query_term.replace("de", "").replace("para", "").replace(".", "").replace("com", "").replace(" e ", ""))
        query_size_results[size] = query_list
        
query_size = {}
for size in query_size_results:
    query_list = query_size_results[size]
    query_size[size] = len(query_list)
```

```python
# show distribution of query size, by token
plt.xlim(0, 20)
new_list = range(0, 20)
plt.xticks(new_list)
plt.bar(list(query_size.keys()), query_size.values(), color='b')
plt.show()
```

```python
# create lists from tokens positions

first_tokens = []
second_tokens = []
third_tokens = []
for size in query_size_results:
    for token in query_size_results[size]:
        tokens = nltk.word_tokenize(token)
        if len(tokens) > 2:
            first_tokens.append(tokens[0])
            second_tokens.append(tokens[1])
        if len(tokens) > 3:
            third_tokens.append(tokens[2])
```

```python
tokens_quantity = {}

for token in first_tokens:
    if token in tokens_quantity:
        tokens_quantity[token] = tokens_quantity[token] + 1
    else:
        tokens_quantity[token] = 1

for token in second_tokens:
    if token in tokens_quantity:
        tokens_quantity[token] = tokens_quantity[token] + 1
    else:
        tokens_quantity[token] = 1
        
for token in third_tokens:
    if token in tokens_quantity:
        tokens_quantity[token] = tokens_quantity[token] + 1
    else:
        tokens_quantity[token] = 1
```
